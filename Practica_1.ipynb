## Previo de la Práctica 1

- Definición de URDF.

    URDF significa Unified Robot Description Format (Formato Unificado de Descripción de Robots). Es un archivo con formato XML que se utiliza en ROS y ROS 2 para representar el modelo físico completo de un robot.

    Describe la geometría visual (cómo se ve), la geometría de colisión (qué partes chocan), las propiedades inerciales (masa y centros de gravedad) y la relación cinemática entre los diferentes componentes del robot.

- Tipos de juntas que están definidas dentro del formato URDF.

    El formato URDF permite especificar 6 tipos fundamentales de juntas:

    * Revolute (Revoluta): Una junta rotacional que se mueve alrededor de un eje y tiene límites de ángulo máximos y mínimos (por ejemplo, el codo de un brazo robótico).

    * Continuous (Continua): Una junta rotacional idéntica a la de revoluta, pero sin límites de giro; puede rotar indefinidamente (por ejemplo, las ruedas de un robot móvil).

    * Prismatic (Prismática): Una junta de deslizamiento lineal que se mueve a lo largo de un eje y posee límites de distancia mínimos y máximos (por ejemplo, un eje telescópico o un pistón).

    * Fixed (Fija): No permite ningún tipo de movimiento. Se usa para soldar o unir rígidamente dos componentes (por ejemplo, fijar una cámara al chasis del robot).

    * Floating (Flotante): Permite el movimiento en los 6 grados de libertad del espacio tridimensional (3 traslaciones y 3 rotaciones). No tiene un eje fijo.

    * Planar (Planar): Permite el movimiento en un plano bidimensional (traslación en X e Y, y rotación en Z).

- Definición de elementos "primitivos" que se pueden utilizar para definir un robot en ROS.

    Al modelar un robot en URDF sin necesidad de cargar un archivo CAD pesado, ROS permite definir la geometría visual y de colisión utilizando cuerpos geométricos básicos conocidos como primitivos:

    * Box (Caja/Cubo): Se define mediante sus tres dimensiones de escala: largo, ancho y alto (x,y,z).

    * Cylinder (Cilindro): Se define especificando únicamente su radio (radius) y su longitud o altura (length).

    * Sphere (Esfera): Se define mediante un solo parámetro: su radio (radius).

- Que es un nodo y que es un topico en ROS 2

    * Nodo (Node): Un nodo es un proceso ejecutable individual diseñado para realizar una tarea específica dentro del robot. La arquitectura de ROS se basa en dividir un sistema complejo en muchos nodos pequeños que cooperan entre sí.

    * Tópico (Topic): Es el canal de comunicación unidireccional y asíncrono que utilizan los nodos para intercambiar información bajo el modelo de Publicador/Suscriptor. Un nodo "publica" datos  en un tópico específico  y cualquier otro nodo que necesite esa información se "suscribe" a dicho tópico para escucharla en tiempo real, sin que los nodos tengan que conocer la existencia del otro.

- Que es RVIZ.

    * RVIZ es un visor de datos del sistema, su función principal es mostrarle al usuario lo que el robot "está viendo y pensando". Se puede visualizar el modelo URDF del robot, las lecturas de sensores láser (LiDAR), nubes de puntos de cámaras 3D, mapas generados, transformaciones cinemáticas (TF) y las trayectorias planeadas, todo renderizado de manera precisa en un entorno tridimensional interactivo.

## Desarrollo de la práctica 

1. La primera parte de realizará en clase, en la cual se desarrollará y simulará de descripción de un eslabón cayo movimiento está restringido por una junta rotacional.

   Se inició con la exploración y simulación en entorno gráfico de la descripción estructural de un robot manipulador base. En esta etapa, el sistema se configuró para validar el comportamiento físico y geométrico de un único eslabón, cuyo rango de movimiento se encuentra estrictamente limitado por una sola junta rotacional.

2. El alumno debe modificar el código del archivo para agregar otro eslabón el cual puede estar unido por al primer eslabón por una junta revolución o una junta prismática.

    Se desarrolló un paquete de ROS 2 en Python (p0_py) y se programó un nodo enfocado en el cálculo e inyección de trayectorias modulares continuas. Al compilar con colcon y ejecutar el nodo, se enviaron señales síncronas a los tópicos de cada junta (/joint1/cmd_pos, /joint2/cmd_pos, etc.), logrando coordinar el movimiento suave del brazo completo y validando la extensión del modelo mediante herramientas de diagnóstico en tiempo real.
