# Práctica 1 Simulación del robot Scara en Gazebo 2

## Objetivo

El alumno deberá cargar y visualizar el modelo virtual de un robot de 3 grados de libertad en el simulador Gazebo 2 para verificar la integridad de su geometría y arquitectura física en un estado estacionario (sin movimiento).


### Metas 

- Dominar el uso de archivos Xacro para modelar la estructura de un robot de 3 eslabones.

- Programar nodos publicadores en Python para transmitir el estado y las transformadas del manipulador.

- Integrar ROS 2 con Gazebo 2 para cargar y visualizar el entorno de simulación.

- Validar visualmente la geometría del robot en estado estático dentro del simulador.





## Previo de la Práctica 1

- Modelado con Xacro:
    Xacro es un lenguaje de macros basado en XML, en si, no es un formato de descripción de robots nuevo en sino un preprocesador creado específicamente para mejorar y simplificar el formato URDF (Unified Robot Description Format).
    Xacro es como usar un lenguaje de programación moderno que te permite usar variables, funciones y condicionales para luego generar ese archivo plano automáticamente.
    Los archivos Xacro evitan la redundancia ya que se crea una "función" (macro) para un eslabón genérico y simplemente la "llamas" las veces necesarias pasándole las dimensiones de cada uno.
    Además permite definir constantes al inicio del documento, como la masa de los eslabones, los límites de los motores o las longitudes, en este caso solo se definieron el tamaño de los eslabones.
    Y en lugar de calcular a mano con una calculadora dónde está el centro de masa de un eslabón o el desfase de un eje, puedes escribir expresiones como xyz="0 0 ${longitud_eslabon / 2}"
    [1]
- Joint y eslabón:
    Eslabón: Representa la parte rígida y física del robot.
    Joint:Representa la unión cinemática entre dos eslabones.
    Juntando ambas funciones, el sistema funciona dividiendo la información del robot en tres capas fundamentales e independientes: 
    
        * La visual, que define la geometría y apariencia del manipulador para su renderizado en pantalla.
        * La física, que especifica las masas, centros de gravedad y propiedades de colisión necesarias para que el motor de simulación aplique leyes dinámicas reales.
        * La matemática, que establece los ejes cinemáticos y transformaciones espaciales para que ROS calcule de manera automática la posición del robot en el espacio tridimensional.

2. Configuración del Launch File en Python 
import os
from ament_index_python.packages import get_package_share_directory
from launch import LaunchDescription
from launch.actions import ExecuteProcess
from launch_ros.actions import Node
from launch.substitutions import Command

def generate_launch_description():

    pkg_scara_description = get_package_share_directory('scara_description')
    pkg_scara_bringup = get_package_share_directory('scara_bringup')

    urdf_path = os.path.join(pkg_scara_description, 'urdf', 'gz2_scara.xacro')
    gazebo_config_path = os.path.join(pkg_scara_bringup, 'config', 'gz_bridge.yaml')
    world_path = os.path.join(pkg_scara_bringup, 'worlds', 'world_config.world')

    rviz_config_path = os.path.join(pkg_scara_bringup, 'rviz', 'scara_config.rviz')

    use_sim_time = {'use_sim_time': True}

    robot_state_publisher_node = Node(
        package='robot_state_publisher',
        executable='robot_state_publisher',
        parameters=[{'robot_description': Command(['xacro ', urdf_path])}, use_sim_time]
    )


    gz_ros_bridge_node = Node(
        package='ros_gz_bridge',
        executable='parameter_bridge',
        parameters=[{'config_file': gazebo_config_path}, use_sim_time],
        output='screen'
    )

    gz_sim = ExecuteProcess(
        cmd=['gz', 'sim', '-r', world_path, '--render-engine', 'ogre'],
        output='screen'
    )

    spaw_entity = Node(
    package='ros_gz_sim',
    executable='create',
    arguments=[
        '-topic', 'robot_description',
        '-name', 'robot_scara',
        '-world', 'world_config' # DEBE coincidir con el <world name="..."> de tu archivo .world
    ],
    output='screen'
    )

    rviz_node = Node(
        package='rviz2',
        executable='rviz2',
        name='rviz2',
        output='screen',
        arguments=['-d', rviz_config_path],
        parameters=[use_sim_time] # Fundamental para evitar errores de TF
    )

    return LaunchDescription([
        robot_state_publisher_node,
        gz_sim,
        spaw_entity,
        gz_ros_bridge_node,
        rviz_node
    ])

3. Integración y Spawning del Robot en Gazebo 2 

Ir al directorio del espacio de trabajo:
    
    cd ~/ROSDev/wssem26

    Compilar el paquete específico del proyecto:
    Bash

    colcon build --packages-select p0_py --symlink-install

    Cargar el entorno para que la terminal reconozca los comandos de tu robot:
    
    source install/setup.bash

    Lanzar Gazebo con el robot SCARA:
    Bash

    ros2 launch scara_bringup gz2_scara.launch.py

    (Al dar Enter aquí, se abrirá la ventana gráfica de Gazebo mostrando el brazo robótico SCARA de 3 juntas).
