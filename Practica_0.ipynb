## Previo de la Práctica 0

Se le solicita al alumno investigar y colocar la sigueinte información en esta sección:

- La estructura de paquetes en ROS 2.

    El desarrollo de sistemas robóticos modernos exige una organización modular que permita la escalabilidad y el trabajo colaborativo. En el ecosistema de ROS 2, esta unidad organizativa es el "paquete" y un paquete es el contenedor más pequeño en el que se puede construir y distribuir software, permitiendo que la comunidad comparta algoritmos y herramientas de manera eficiente.

    La estructura de un paquete en ROS 2 varía ligeramente dependiendo del lenguaje de programación y del sistema de construcción utilizado. Sin embargo, existen elementos universales. El archivo package es el corazón de los metadatos; en él se definen el nombre del paquete, la versión, los autores y, lo más importante, las dependencias de otros paquetes de ROS 2.

    Dentro de la carpeta del paquete, encontramos directorios específicos: src para el código fuente, include para los encabezados, y launch para archivos que permiten ejecutar múltiples nodos con un solo comando. Además, si el paquete define datos personalizados, se incluyen carpetas como msg (mensajes), srv (servicios) o action. Esta arquitectura jerárquica asegura que cualquier desarrollador pueda navegar por el código de terceros y entender rápidamente dónde reside cada componente del sistema.

- Definición de Nodos de Tópicos.
    La arquitectura de software en robótica ha evolucionado hacia sistemas distribuidos donde el procesamiento no ocurre en un solo bloque monolítico. En ROS 2, esta filosofía se materializa a través de los conceptos de Nodos y Tópicos. Estos elementos forman el "Grafo de ROS", una red de procesos interconectados que permite que un robot perciba su entorno, procese la información y actúe en consecuencia de manera coordinada y eficiente.

    Un Nodo es un proceso especializado que realiza una tarea específica dentro del sistema robótico. Esta modularidad permite que, si un nodo falla, el resto del sistema pueda seguir funcionando o reiniciarse de forma aislada. Funcionan bajo un modelo de comunicación de "Publicador/Suscriptor". Un nodo publica información en un tópico y cualquier otro nodo interesado puede suscribirse a ese tópico para recibir la información en tiempo real. Esta comunicación es unidireccional y asíncrona, lo que significa que el nodo que envía la información no necesita esperar una respuesta del que la recibe.

    Al desacoplar la producción de datos de su consumo, los desarrolladores pueden añadir nuevas funcionalidades simplemente suscribiéndose a tópicos existentes sin necesidad de modificar el código original.



## Desarrollo de la práctica 

El desarrollo de la práctica se divide en dos partes:

1. Conocer las herramientas computacionales para la robótica de la clase.

    Se configuró y exploró el entorno operativo Linux (Ubuntu) y la distribución ROS 2 Jazzy. Se utilizaron comandos esenciales de terminal (cd, chmod, source) para la navegación del sistema de archivos, y se empleó Visual Studio Code como entorno de desarrollo integrado (IDE) principal para la edición del código fuente.

2. Generación de un paquete de ROS 2 en Python.

    Se estructuró un paquete de software nativo utilizando la herramienta de línea de comandos de ROS 2 mediante el tipo de construcción ament_python. Este paso automatizó la creación de los archivos de configuración esenciales (package.xml y setup.py), definiendo las dependencias del proyecto (rclpy, std_msgs) y estableciendo una arquitectura limpia para el manejo de nodos.

3. Desarrollo de los Nodos y Tópicos correspondientes para la práctica.

    Se programó un nodo publicador orientado a objetos (MyNode) que calcula de forma analítica trayectorias armónicas suaves para las articulaciones de un robot manipulador. Asimismo, se configuró el archivo setup.py dentro de la sección entry_points para registrar formalmente los scripts en el sistema, permitiendo su compilación mediante colcon build y su posterior ejecución en terminal con ros2 run.

4. El manejo de herraminetas para la documentación empleando Jupyter Lab.

    Se utilizó el entorno interactivo de Jupyter Lab como la herramienta central para documentar los flujos de trabajo en celdas de Markdown y código. Se integraron bitácoras de terminal y capturas de pantalla del monitoreo de tópicos en tiempo real (ros2 topic echo), garantizando un reporte reproducible que valida conceptual y prácticamente los resultados obtenidos.

### Conclusiones


La implementación del nodo publicador no solo cumplió con el objetivo de establecer un canal de comunicación activo, sino que demostró una aplicación avanzada de la robótica al transmitir de forma síncrona y precisa (a una frecuencia de 50 Hz) los perfiles de trayectoria armónica modular para las juntas de un manipulador SCARA. La correcta recepción de estas señales a través del comando ros2 topic echo confirma que la arquitectura distribuida funciona de manera robusta y transparente.

Se pudo constatar que la integración de Visual Studio Code para la codificación, la terminal de comandos para la compilación con colcon, y Jupyter Lab para la documentación interactiva, constituye un buen flujo de trabajo y que además es indispensable para el diseño, depuración y reporte de sistemas robóticos complejos en el ámbito profesional y de investigación.
