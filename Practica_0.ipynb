## Previo de la Práctica 0

- La estructura de paquetes en ROS 2.

    - El desarrollo de sistemas robóticos modernos exige una organización modular que permita la escalabilidad y el trabajo colaborativo. En el ecosistema de ROS 2, esta unidad organizativa es el "paquete" y un paquete es el contenedor más pequeño en el que se puede construir y distribuir software, permitiendo que la comunidad comparta algoritmos y herramientas de manera eficiente. 
    
        La estructura de un paquete en ROS 2 varía ligeramente dependiendo del lenguaje de programación y del sistema de construcción utilizado. Sin embargo, existen elementos universales. El archivo package  es el corazón de los metadatos; en él se definen el nombre del paquete, la versión, los autores y, lo más importante, las dependencias de otros paquetes de ROS 2.

        Dentro de la carpeta del paquete, encontramos directorios específicos: src para el código fuente, include para los encabezados, y launch para archivos que permiten ejecutar múltiples nodos con un solo comando. Además, si el paquete define datos personalizados, se incluyen carpetas como msg (mensajes), srv (servicios) o action. Esta arquitectura jerárquica asegura que cualquier desarrollador pueda navegar por el código de terceros y entender rápidamente dónde reside cada componente del sistema.
- Definición de Nodos de Tópicos.
    - La arquitectura de software en robótica ha evolucionado hacia sistemas distribuidos donde el procesamiento no ocurre en un solo bloque monolítico. En ROS 2, esta filosofía se materializa a través de los conceptos de Nodos y Tópicos. Estos elementos forman el "Grafo de ROS", una red de procesos interconectados que permite que un robot perciba su entorno, procese la información y actúe en consecuencia de manera coordinada y eficiente.

        Un Nodo es un proceso especializado que realiza una tarea específica dentro del sistema robótico. Esta modularidad permite que, si un nodo falla, el resto del sistema pueda seguir funcionando o reiniciarse de forma aislada. 
    Funcionan bajo un modelo de comunicación de "Publicador/Suscriptor". Un nodo publica información en un tópico y cualquier otro nodo interesado puede suscribirse a ese tópico para recibir la información en tiempo real. Esta comunicación es unidireccional y asíncrona, lo que significa que el nodo que envía la información no necesita esperar una respuesta del que la recibe.
    
        Al desacoplar la producción de datos de su consumo, los desarrolladores pueden añadir nuevas funcionalidades simplemente suscribiéndose a tópicos existentes sin necesidad de modificar el código original.
